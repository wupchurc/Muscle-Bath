```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

```

```python
df= pd.read_csv('./2018_08_22_Myometrial_Contractility_Study_4_MB_1_8.tsv', sep='\t')
df.drop(columns='Timestamp')
df.drop(columns='Stimulus')
df.set_index('Experiment Time', inplace=True)
inds_base = df.index[df['Comments'] == 'Baseline']
inds_treat = df.index[df['Comments'] == 'Treatment Bath 1']

df_base = df[inds_base[0]:inds_treat[0]]
df_treat = df[inds_treat[0]:]
```

```python
#plotting graphs for Muscle Bath 1
df_base[['Force1']].plot(title = 'Baseline')
df_treat[['Force1']].plot(title = 'Treatment')
```

```python
#determine peak force by scipy.signal function
peak_indices = signal.argrelmax(np.array(df_base.Force1),order =500)


print(peak_indices)
#print(np.array(df_base.Force1)[peak_indices])
print(df_base.Force1.iloc[peak_indices])
peak_count = len(peak_indices)
```
